### Character Encoding

Computers have different ways of representing and interpreting characters.

Sometimes, we have to specify the encoding to use otherwise our code can not process the text.

```python
import nltk

di = open("dec_independence.txt").read()

print(di[:100])
```

```python
di_token = nltk.word_tokenize(di)

di_freq = nltk.FreqDist(di_token)

print(di_freq.most_common(10))
```

### HTML

```python
from urllib.request import urlopen

url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

html = urlopen(url).read()

print(html[:100])
```

```python
from bs4 import BeautifulSoup

web_str = BeautifulSoup(html).get_text()

print (web_str)
```

```python
start = web_str.find("Python is a high-level, interpreted, general-purpose programming language.")
end = web_str.find("Python's syntax allows specifying static types, but they are not checked in the default implementation, CPython.")
last_sent = len("Python's syntax allows specifying static types, but they are not checked in the default implementation, CPython.")

intro = web_str[start:end+last_sent]

intro_tokens = nltk.word_tokenize(intro)
print(intro_tokens[:20])
```

### Directly from URL

```python
#Bringing in text directly from a URL

new_url = "http://www.gutenberg.org/cache/epub/26275/pg26275.txt"

odyssey_str = urlopen(new_url).read().decode("utf-8")

print("odyssy_str:", odyssey_str[5:47])
```

```python
odyssey_tokens = nltk.word_tokenize(odyssey_str)

print("odyssey_tokens:", odyssey_tokens[:20])
```

```python
print("Length of odyssey_tokens:", len(odyssey_tokens))

print("type of odyssey_tokens:", type(odyssey_tokens))
```

```python
#Turning into nltk data structure

odyssey_text = nltk.Text(odyssey_tokens)
```

### CSV

```python
import csv

comments = []

with open("../Mini Projects/Sentiment Analysis/reviews.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)

print("len comments: ", len(comments))
print("first comment: ", comments[0])
```

```python
comment_tokens = [nltk.word_tokenize(str(entry)) for entry in comments]

print("comment_tokens: ", comment_tokens[0][:5])
```

### Exporting

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")

print("Alice: ", alice[:100])
print("Alice length: ", len(alice))
```

```python
alice_str = ' '.join(alice)

print(alice_str[:100])
```

```python
export_file = open("alice_exported.txt", "w")

export_file.write(alice_str)

export_file.close()
```

### NLTK RESOURCES

```python
#English words list - think of as all the words that may be found in an english dictionary.
english_words = nltk.corpus.words.words()[456:1000]

print("english_words: ", english_words[:10])
```

```python
#List of stop words
stop_words = nltk.corpus.stopwords.words("english")

print("Stop words: ", stop_words[:10])
```

### Example - Remove stop words

Task: Remove stop words from the 100 most common words in Alice in Wonderland

```python
alice_fd = nltk.FreqDist(alice)

alice_100_words = alice_fd.most_common(1000)

alice_100_words = [(word.lower(), count) for word, count in alice_100_words]

print("alice_100_words: ", alice_100_words[:10])
print("alice_100_words length: ", len(alice_100_words))

alice_100_words_no_stop = [word for word in alice_100_words if word[0] not in stop_words and word[0].isalpha()]

print("alice_100_words_no_stop: ", alice_100_words_no_stop)
print("alice_100_words_no_stop length: ", len(alice_100_words_no_stop))
```
